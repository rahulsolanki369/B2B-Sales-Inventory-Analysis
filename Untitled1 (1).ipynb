## Missing Value Analysis
Identify missing values in the dataset.

```python
# Fill Missing Values

df["Stock in Hand"] = df["Stock in Hand"].fillna(0)
df["Unit Price"] = df["Unit Price"].fillna(0)
df["Total Value"] = df["Total Value"].fillna(0)
```

## Data Cleaning
Replace missing values with appropriate values.

```python
# Verify Missing Values

df.isnull().sum()
```

## Top 5 Products by Total Value
Identify the products with the highest inventory value.

```python
# Top 5 Products by Total Value

top_products = df.sort_values(by="Total Value",ascending=False)

top_products.head(5)
```

## Category-wise Analysis
Analyze total inventory value by product category.

```python
# Category-wise Total Value Analysis

category_value = df.groupby("Category Name")["Total Value"].sum()
category_value
```

## Data Visualization
Visualize the top five products using a bar chart.

```python
# Top 5 Products Bar Chart

import matplotlib.pyplot as plt

top5 = df.sort_values(by="Total Value", ascending=False).head(5)

plt.figure(figsize=(10,5))
plt.bar(top5["Item Name"], top5["Total Value"])
plt.title("Top 5 Products by Total Value")
plt.xlabel("Item Name")
plt.ylabel("Total Value")
plt.xticks(rotation=45)
plt.show()
```

## Highest Stock Items
Display the products with the highest available stock.

```python
# Top 5 Stock Items

category_value = df.groupby("Category Name")["Total Value"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_value.plot(kind="bar")
plt.title("Category-wise Total Value")
plt.xlabel("Category")
plt.ylabel("Total Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
